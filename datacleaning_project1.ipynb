```python
import os
import pandas as pd


#read the casv file
df = pd.read_csv('pancreatic_cancer_prediction_sample.csv', nrows=10000)

print("Preview of the first dataset")
display(df.head())

df = df.drop_duplicates()

#Filter for cancer patients under age 40
result = df[df['Age'] < 40]

# Display result
display(result)


count_under40 = len(result)
print(f"Total number of cancer patients under 40: {count_under40}")

total_patients = df.shape[0]
print(f"Total number of cancer patients: {total_patients}")

result1 = df[df['Obesity'] == 1]

sum = len(result1)
print(f" Number  of patients who were obese: {sum}")

pertage = ((sum/total_patients)* 100)
print(f" Percentage of people who were obese: {pertage} % ")

df.to_csv('cleaned_data_pancreatic_cancer_prediction.csv', index=False)


```
